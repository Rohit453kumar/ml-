**Ensemble Techniques**

- Bagging - Building multiple models (typically of the same type) from different subsamples of the training dataset.
- Boosting - Building multiple models (typically of the same type) each of which learns to fix the prediction errors of a prior model in the chain.
- Voting - Building multiple models (typically of differing types) and simple statistics (like calculating the mean) are used to combine predictions.

```python
# check missing values in the training dataset
missing_values=train_data.isnull().sum()
missing_values=missing_values[missing_values>0].sort_values(ascending=False)
missing_values
```

```python
# check the percentage of null values
null_values_per_column=train_data.isnull().sum()/len(train_data)*100
per_column=null_values_per_column[null_values_per_column>0].sort_values(ascending=False)
per_column
```

There are three rows which have more than 90% values are null so we remove those featues and we impute the remaining features by using median for numerical columns and mode for categorical columns

for column fence we would check if it is highly correlated with target column saleprices then we keep the column or we remove it

```python
from scipy.stats import chi2_contingency

# Create a contingency table for `Fence` and `SalePrice`
contingency_table = pd.crosstab(train_data['Fence'], train_data['SalePrice'] > train_data['SalePrice'].median())
chi2, p, _, _ = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("p-value:", p)
```

```python
if p>0.05:
    print("they are no highly correlated")
else:
    print("column saleprice is highly dependent or correlated with fenced")
```

```python
# Test the variance of sales price with different fence groups
from scipy.stats import f_oneway

# Group SalePrice by Fence categories
fence_groups = [group['SalePrice'].dropna() for _, group in train_data.groupby('Fence')]
anova_result = f_oneway(*fence_groups)

print("ANOVA Statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

```python
if anova_result.pvalue>0.05:
    print("salePrice is the same across all Fence categories.")
else:
    print("SalePrice differs significantly across the different Fence categories")
```

```python
# visualize fence vs saleprice with the help of boxplot
sns.boxplot(x='Fence', y='SalePrice', data=train_data)
plt.title("SalePrice vs. Fence")
plt.xticks(rotation=45)
plt.show()
```

since column fence is correlated hence we do not remove it

```python
# drop the columns which have more than 90% null values
columns_to_drop = ['PoolQC', 'MiscFeature', 'Alley']
train_data_cleaned = train_data.drop(columns=columns_to_drop)
```

```python
train_data_cleaned
```

```python
#Impute the numerical column with median
from sklearn.impute import SimpleImputer
numerical_columns = train_data_cleaned.select_dtypes(include=['float64', 'int64']).columns
imputer_num = SimpleImputer(strategy='median')
train_data_cleaned[numerical_columns] = imputer_num.fit_transform(train_data[numerical_columns])
```

```python
# impute the categorical columns using the mode
categorical_columns = train_data_cleaned.select_dtypes(include=['object']).columns
imputer_cat = SimpleImputer(strategy='most_frequent')
train_data_cleaned[categorical_columns] = imputer_cat.fit_transform(train_data_cleaned[categorical_columns])
```

```python
train_data_cleaned
```

```python
missing_val=train_data_cleaned.isnull().sum()
missing_val=missing_val[missing_val>0].sort_values(ascending=False)
missing_val
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    train_data_cleaned[col] = le.fit_transform(train_data[col])
    label_encoders[col] = le
```

```python
corr_matrix=train_data_cleaned.corr()
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
#now we check the scatter plot for lot area vs sale price column
sns.scatterplot(x='LotArea', y='SalePrice', data=train_data_cleaned, color="green")
plt.title("LotArea vs. SalePrice")
plt.xlabel("LotArea")
plt.ylabel("SalePrice")
plt.show()
```

```python
# split the train dataset into training sets and validation sets
from sklearn.model_selection import train_test_split
x = train_data_cleaned.drop(columns=['SalePrice', 'Id'])
y = train_data_cleaned['SalePrice']


x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)

# Output dataset shapes for verification
x_train.shape, x_val.shape, y_train.shape, y_val.shape
```

```python
# train and predict by decision tree regressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,accuracy_score,precision_score,recall_score,f1_score
#Train the model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(x_train, y_train)
```

```python
# make prediction
y_pred_dt = dt_model.predict(x_val)
```

```python
#  Evaluate the model
mae_dt = mean_absolute_error(y_val, y_pred_dt)
rmse_dt = np.sqrt(mean_squared_error(y_val, y_pred_dt))
r2_dt = r2_score(y_val, y_pred_dt)*100


mae_dt, rmse_dt, r2_dt
```

```python
# random forest regressor model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,accuracy_score,precision_score,recall_score,f1_score
# train the model
rf_model=RandomForestRegressor(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
```

```python
# make prediction
y_pred_rf=rf_model.predict(x_val)
```

```python
# Evalaute the model
mae_rf = mean_absolute_error(y_val, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_val, y_pred_rf))
r2_rf = r2_score(y_val, y_pred_rf)*100

mae_rf,rmse_rf,r2_rf
```

```python
#Ada boost regressor model
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor(n_estimators=100, random_state=42)
ada.fit(x_train, y_train)
```

```python
y_pred_ada=ada.predict(x_val)
```

```python
# Evalaute the model
mae_ada = mean_absolute_error(y_val, y_pred_ada)
rmse_ada = np.sqrt(mean_squared_error(y_val, y_pred_ada))
r2_ada = r2_score(y_val, y_pred_ada)*100
mae_ada,rmse_ada,r2_ada
```

```python
# gradient boosting regression
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(n_estimators=100, random_state=42)
gbr.fit(x_train, y_train)
```

```python
y_pred_gbr=gbr.predict(x_val)
```

```python
mae_gbr = mean_absolute_error(y_val, y_pred_gbr)
rmse_gbr = np.sqrt(mean_squared_error(y_val, y_pred_gbr))
r2_gbr = r2_score(y_val, y_pred_gbr)*100
mae_gbr,rmse_gbr,r2_gbr
```

```python
# XG boost regression
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=100, random_state=42, objective='reg:squarederror')
xgb.fit(x_train, y_train)
```

```python
y_pred_xgb=xgb.predict(x_val)
```

```python
# Evaluate the model
mae_xgb = mean_absolute_error(y_val, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_val, y_pred_xgb))
r2_xgb = r2_score(y_val, y_pred_xgb)*100
mae_xgb,rmse_xgb,r2_xgb
```

```python
#organize the metrics
evaluation_table = pd.DataFrame({
    "Model": ["DecisionTreeRegressor","RandomForestRegressor","AdaBoostRegressor","GradientBoostingRegressor","XGBRegressor"],
    "MAE":[mae_dt,mae_rf,mae_ada,mae_gbr,mae_xgb],
    "RMSE":[rmse_dt,rmse_rf,rmse_ada,rmse_gbr,rmse_xgb],
    "R2-Score":[r2_dt,r2_rf,r2_ada,r2_gbr,r2_xgb],
})
evaluation_table
```
